```python
import numpy as np
import matplotlib.pyplot as plt
```

Função constante

f(x) = c

Exemplo 1: Considere a velocidade constante de um carro de 30km/h

```python
fig = plt.figure(figsize=(8,5))
plt.title('Função constante')

x = np.linspace(0, 10, 10000)
y= 30*np.ones(x.shape)

plt.plot(x, y)
plt.show()
```

Função linear

f(x) = ax + b

Exemplo 2: Preço a ser cobrado numa corrida de taxi, considerando o b=10 o valor fixo do serviço e a=2 o valor para cada quilômetro percorrido.

```python
fig = plt.figure(figsize=(8,5))
plt.title('Função linear')

x = np.linspace(0, 10, 10000)
y= 5 + 2*x

plt.plot(x, y)
plt.show()
```

Função quadrática

f(x) = ax² + bx + c

Exemplo 3: Trajetória de uma bola lançada considerando o efeito da gravidade

```python
fig = plt.figure(figsize=(8,5))
plt.title('Função quadrática')

x = np.linspace(0, 4, 10000)
y= 4 -(x-2)**2

plt.plot(x, y)
plt.show()
```

Função racional

```python
fig = plt.figure(figsize=(8,5))
plt.title('Função racional')

x = np.linspace(0.1, 0.5, 10000)
y= (5e-15)*((1/x)**12 - (9.6/x)**6)

plt.plot(x, y)
plt.show()
```

Função seno

```python
fig = plt.figure(figsize=(8,5))
plt.title('Função seno')

x = np.linspace(-3.1415, 3.1415, 10000)
y= np.sin(x)

plt.plot(x, y)
plt.show()
```

Função cosseno

```python
fig = plt.figure(figsize=(8,5))
plt.title('Função cosseno')

x = np.linspace(-3.1415, 3.1415, 10000)
y= np.cos(x)

plt.plot(x, y)
plt.show()
```

Função tangente

```python
fig = plt.figure(figsize=(8,5))
plt.title('Função tangente')

x = np.linspace(-1, 1, 10000)
y= np.tan(x)

plt.plot(x, y)
plt.show()
```

Função exponencial

Exemplo: número de proliferação de bactérias numa amostra. Considere c=2000 o número inicial de bactérias e taxa de crescimento r=5.

```python
fig = plt.figure(figsize=(8,5))
plt.title('Função exponencial')

x = np.linspace(-1, 1, 10000)
y= 2000 + np.exp(5*x)

plt.plot(x, y)
plt.show()
```

Função sigmóide

```python
fig = plt.figure(figsize=(8,5))
plt.title('Função seno')

x = np.linspace(-4, 4, 10000)
y= 1/(1 + np.exp(-x))

plt.plot(x, y)
plt.show()
```
