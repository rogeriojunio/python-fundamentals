```python
import numpy as np
import matplotlib.pyplot as plt
```

Função constante

f(x) = c

Exemplo 1: Considere a velocidade constante de um carro de 30km/h

```python
fig = plt.figure(figsize=(8,5))
plt.title('Função constante')

x = np.linspace(0, 10, 10000)
y= 30*np.ones(x.shape)

plt.plot(x, y)
plt.show()
```

Função linear

f(x) = ax + b

Exemplo 2: Preço a ser cobrado numa corrida de taxi, considerando o b=10 o valor fixo do serviço e a=2 o valor para cada quilômetro percorrido.

```python
fig = plt.figure(figsize=(8,5))
plt.title('Função linear')

x = np.linspace(0, 10, 10000)
y= 5 + 2*x

plt.plot(x, y)
plt.show()
```

Função quadrática

f(x) = ax² + bx + c

Exemplo 3: Trajetória de uma bola lançada considerando o efeito da gravidade

```python
fig = plt.figure(figsize=(8,5))
plt.title('Função quadrática')

x = np.linspace(0, 4, 10000)
y= 4 -(x-2)**2

plt.plot(x, y)
plt.show()
```
